A line chart or line plot is a type of plot which displays information as a series of data points called 'markers' connected by straight line segments. It is a basic type of chart common in many fields.
Use line plot when you have a continuous data set. These are best suited for trend-based visualizations of data over a period of time.

```python
df_top5.index = df_top5.index.map(int)
df_top5.plot(kind='area',
             stacked=False,
             alpha=0.25,         # transparency, default=0.5
             figsize=(20, 10))

plt.title('Immigration Trend of Top5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

### Two types of plotting

There are two styles/options of ploting with `matplotlib`. Plotting using the Artist layer and plotting using the scripting layer.

**Option 1: Scripting layer (procedural method) - using matplotlib.pyplot as 'plt'**

You can use `matplotlib.pyplot` and add more elements by calling different methods procedurally; for example, `plt.title(...)` to add title or `plt.xlabel(...)` to add label to the x-axis.

**Option 2: Artist layer (Object oriented method) - using an `Axes` instance from Matplotlib (preferred)**

You can use an `Axes` instance of your current plot and store it in a variable (eg. `ax`). You can add more elements by calling methods with a little change in syntax (by adding "*set_*" to the previous methods). For example, use `ax.set_title()` instead of `plt.title()` to add title,  or `ax.set_xlabel()` instead of `plt.xlabel()` to add label to the x-axis. 

This option sometimes is more transparent and flexible to use for advanced plots (in particular when having multiple plots, as you will see later). 

In this course, we will stick to the **scripting layer**, except for some advanced visualizations where we will need to use the **artist layer** to manipulate advanced aspects of the plots.

```python
df_bttm5 = df.tail(5)
df_bttm5 = df_bttm5[years].transpose()

df_bttm5.index=df_bttm5.index.map(int)

df_bttm5.plot(kind='area',\
              stacked=False,\
              alpha=0.5,\
              figsize=(20,10))

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

```python
Ctrs = ['Greece', 'Albania', 'Bulgaria']
df_Ctrs=df.loc[Ctrs, years].transpose()
df_Ctrs.head()
```

```python
count, bin_edges = np.histogram(df_Ctrs, 15)
xmin = bin_edges[0] - 10
xmax = bin_edges[-1] +10

df_Ctrs.plot.hist(figsize=(10,6),
                  bins=15,
                  alpha=0.5,
                  xticks=bin_edges,
                  color=['coral', 'darkslateblue', 'mediumseagreen'],
                  stacked=False,
                  xlim=(xmin, xmax))

str_Ctrs = ', '.join(Ctrs)

plt.title('Histogram of Immigration from %s from 1980 - 2013' % str_Ctrs)
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```

### Bar Charts (Dataframe)

```python
ctr = 'Iceland'
df_ctr = df.loc[ctr, years]
df_ctr.head()
```

```python
df_ctr.plot(kind='bar', figsize=(10,6), color='grey', rot=90) # rotate the bars by 90 degrees

plt.title('%s immigrants to Canada from 1980 to 2013' % ctr)
plt.ylabel('Number of immigrants')
plt.xlabel('Years')

# Annotate arrow
plt.annotate('',                      # s: str. Will leave it blank for no text
             xy=(32, 70),             # place head of the arrow at point (year 2012 , pop 70)
             xytext=(28, 20),         # place base of the arrow at point (year 2008 , pop 20)
             xycoords='data',         # will use the coordinate system of the object being annotated 
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='blue', lw=2)
            )

# Annotate Text
plt.annotate('2008 - 2011 Financial Crisis', # text to display
             xy=(28, 30),                    # start the text at at point (year 2008 , pop 30)
             rotation=72.5,                  # based on trial and error to match the arrow
             va='bottom',                    # want the text to be vertically 'bottom' aligned
             ha='left',                      # want the text to be horizontally 'left' aligned.
            )


plt.show()
```
